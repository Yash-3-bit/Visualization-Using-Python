#### 1.1 Genrating sin and cos graphs

```python
x = np.linspace(0,np.pi,20)
sin_y = np.sin(x)
plt.plot(x,sin_y)
plt.title('sine wave')
plt.show()
```

```python
x = np.linspace(0,np.pi,30)

#defining wave
sin_y = np.sin(x)
cos_y = np.cos(x)

#plotting the wave
plt.plot(x,sin_y, label = 'sine')
plt.plot(x,cos_y,label = 'cosine')

plt.legend()
plt.title('sin wave')

plt.show()
```

```python
x = np.linspace(0,3*np.pi,30)

#defining wave
sin_y = np.sin(x)
cos_y = np.cos(x)

#plotting the wave
plt.plot(x,sin_y,label='sine')
plt.plot(x,cos_y,label='cosine')

plt.legend(loc='upper right')
plt.title('sin wave')

plt.show()
```

```python
# specifing patterns
x = np.linspace(0,3*np.pi,50)

sin_y = np.sin(x)
cos_y = np.cos(x)

plt.plot(x,sin_y, '*')
plt.plot(x,cos_y,'-')

plt.title('sin-cos wave')
plt.show()

#no need of legend() when specify different patterns.
```

## 4.Different types of plots

```python
#line plot

x = np.linspace(0,4*np.pi,100)
y = np.sin(x)
plt.style.use('Solarize_Light2')

plt.plot(x,y)

plt.title('sine-cos wave')
plt.show()

```

### multiple line plots

```python
x = np.linspace(0,4*np.pi,100)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(15,5))
plt.style.use('Solarize_Light2')

plt.plot(x,y_sin,'*',color='red',linewidth=5,label='sine')
plt.plot(x,y_cos, '--', linewidth=2,label='cosine')

plt.title('sine-cos wave')
plt.show()
```

## pie chart

```python
cars = ['audi','bmw','jaguar','tesla']
data = [450,290,140,31]

fig = plt.figure(figsize=(7,7))

plt.pie(data,labels=cars)
plt.show()
```

```python
#basic operations
cars = ['audi','bmw','jaguar','tesla']
data = [450,290,140,31]
explode = [.1,.2,.3,.5]

fig = plt.figure(figsize=(7,7))

plt.pie(data,labels=cars,explode=explode)
plt.show()
```

```python
#shadows

cars = ['audi','bmw','jaguar','tesla']
data = [450,290,140,31]
explode = [.1,.2,.3,.5]

fig = plt.figure(figsize=(7,7))

plt.pie(data,labels=cars,explode=explode,shadow=True)
plt.show()
```

```python
#adding percentage
cars = ['audi','bmw','jaguar','tesla']
data = [450,290,140,31]
explode = [.1,.2,.3,.5]

fig = plt.figure(figsize=(7,7))

plt.pie(data,labels=cars,explode=explode,autopct='%2.1f%%')
plt.show()
```

```python
#adding multiple widgets
fig = plt.figure(figsize=(7,7))

plt.pie(data,labels=cars,explode=explode,autopct='%2.1f%%',
       wedgeprops={'edgecolor':'black','linewidth':2,'antialiased':False})
plt.show()
```

## multiple subplots

```python
x = np.linspace(0,10,50)

fig,ax = plt.subplots(2,2)

y = np.sin(x)
ax[0,0].plot(x,y)
ax[0,0].set_title('sin wave')

y=np.cos(x)
ax[0,1].plot(x,y)
ax[0,1].set_title('cos wave')

y = np.tan(x)
ax[1,0].plot(x,y)
ax[1,0].set_title('tan wave')

y = np.random.randn(100)
ax[1,1].hist(y,bins=40)
ax[1,1].set_title('random distribution')

plt.tight_layout()
plt.show()
```

```python
plt.subplots_adjust(hspace=.4,wspace=.4)
for i in range(1,7):
    plt.subplot(2,3,i)
    plt.text(0.5,0.5,str((2,3,i)),fontsize=15,ha='center')
```

```python
cars = ['audi','bmw','jaguar','tesla']
data = [450,290,140,31]
explode = [0,0,.3,.5]

fig = plt.figure(figsize=(7,7))

plt.pie(data,labels=cars,explode=explode,autopct='%2.1f%%')
plt.show()
```

### with loops

```python
x = np.linspace(0,10,50)
y = np.sin(x)

for i in range(1,4):
    fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(15,5))
    ax1.plot(x,y)
    ax2.plot(x,y**2)
    ax3.plot(x,y**3)

    ax1.set_title('sin wave')
    ax2.set_title('sin wave - square')
    ax3.set_title('sin wave - cube')

    plt.tight_layout()

    plt.show()
```
